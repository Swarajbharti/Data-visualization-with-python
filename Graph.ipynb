```python
#to add x and y labels in a graph
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,100])
y=np.array([0,250])
plt.plot(x,y,marker='o', linestyle='dashed', color='r')
plt.xlabel("average")
plt.ylabel("beats")
plt.show()
```

```python
#to add labels and title in a graph
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,500])
y=np.array([0,500])
plt.plot(x, y, marker='o', linestyle='dotted', color='g')
plt.xlabel("average")
plt.ylabel("beat")
plt.title("sample graph")
plt.show()
```

```python
#to add labels and title on left side
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,500])
y=np.array([0,100])
plt.xlabel("average")
plt.ylabel("beat")
plt.title("sample graph", loc='left')
plt.plot(x, y, marker='o', linestyle='dotted', color='b')
plt.show()
```

```python
#set font properties for titles & labels
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,100])
y=np.array([0,200])
font1={'family':'serif', 'color':'blue', 'size':15}
font2={'family':'serif', 'color':'red', 'size':20}
plt.xlabel("average", fontdict=font1)
plt.ylabel("beat", fontdict=font1)
plt.title("sample graph", loc='left', fontdict=font2)
plt.plot(x,y, marker='o', linestyle='dotted', color='r')
plt.show()
```

```python
#add gridlines in a graph
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,250])
y=np.array([0,500])
font1={'family':'serif','color':'blue', 'size':20}
plt.xlabel("average",fontdict=font1, loc='center')
plt.ylabel("beat", fontdict=font1, loc='center')
plt.title("sample graph", loc='center')
plt.grid()
plt.plot(x,y, marker='o', linestyle='dotted', color='r')
plt.show()
```

```python
#set properties for grid
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,250])
y=np.array([0,250])
font1={'family':'serif', 'color':'darkgreen', 'size': 20}
plt.xlabel("average", loc='center', fontdict=font1)
plt.ylabel("beat", loc='center', fontdict=font1)
plt.title("sample graph", loc='center', fontdict=font1)
plt.grid(color='blue',linestyle='dashed')
plt.plot(x,y, marker='o', color='red',linewidth=1, linestyle='dotted')
plt.show()
```

```python
#set properties for grid
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,3,6,8])
y=np.array([1,4,1,10])
font1={'family':'serif', 'size':20, 'color':'red'}
plt.xlabel("average", fontdict=font1, loc='center')
plt.ylabel("beat",fontdict=font1, loc='center')
plt.title("sample graph", loc='center', fontdict=font1)
plt.plot(x,y, marker='o', linestyle='dotted', linewidth=0.5, color='red')
plt.grid(linestyle='dashed', color='blue',linewidth=1)
plt.show()
```

```python
#to draw sub plots
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,1,2,3])
y=np.array([1,3,5,7])
plt.title("average", loc='center')
plt.subplot(1,2,1)            #plt.subplot(row, column, index number of plot)
plt.plot(x,y)

x=np.array([0,1,2,3])
y=np.array([1,5,7,9])
plt.title("beats", loc='center')
plt.subplot(1,2,2)
plt.plot(x,y)

plt.show()
```

```python
#to draw scatter plot
import matplotlib.pyplot as plt
import numpy as np
x=np.array([2,3,6,7,9,12,15,19,21,24])       #x and y array should be of same size
y=np.array([34,56,67,23,78,90,32,12,56,45])
plt.scatter(x,y)
plt.show()
```

```python
#compare two scatter plots
import matplotlib.pyplot as plt
import numpy as np
x=np.array([2,3,4,5,6,7,8,2,1,5,7,8])
y=np.array([56,56,78,89,90,91,87,67,90,88,77,45])
plt.scatter(x,y)

x=np.array([6,4,3,2,1,5,6,6,7, 2, 5, 6])
y=np.array([78, 67, 89, 90, 45, 34, 56, 67, 88, 90, 34, 54])
plt.scatter(x,y)
plt.show()
```

```python
#to use colormaps
import matplotlib.pyplot as plt
import numpy as np
x=np.array([2,4,6,7,8,2,4])
y=np.array([34,56,78,90,23,45,56])
colors=np.array([0,10,20,30,40,50,60])
plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()
plt.show()
```

```python
#to draw bar graph
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A", "B", "C", "D"])
y=np.array([3, 8, 10, 1])
plt.xlabel('node')
plt.ylabel('chart')
plt.title('sample')
plt.bar(x,y)
plt.show()
```

```python
#to draw horizontal graphs
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A", "B", "C", "D"])
y=np.array([2, 5, 8, 9])
plt.barh(x,y)
plt.show()
```

```python
#draw histograms
import matplotlib.pyplot as plt
import numpy as np
x=np.random.normal(170,10, 250)   #values will be around 170, standard deviation is 10 and 250 random values
plt.hist(x)
plt.show()
```

```python
#creating piechart
import matplotlib.pyplot as plt
import numpy as np
label=["fronx", "grand vitara", "i20", "nexon", "creta"]
data= [25,50,20,10,29]
fig=plt.figure(figsize=(10,7))
plt.pie(data, labels=label)
plt.show()
```

```python
#to draw boxplot
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(10)                        #creating dataset
data = np.random.normal(100, 20, 200)
fig = plt.figure(figsize =(10, 7))        #length and height of boxplot
plt.boxplot(data)
plt.show()
```

```python
#show outliers using boxplot
import matplotlib.pyplot as plt
df=[10,15,22,330,30,45,60]
plt.boxplot(df,)
plt.title("Box Plot to detect Outliers")
```

```python
#to draw heatmap (dark means higher concentration and light means less concentration)
import matplotlib.pyplot as plt
import numpy as np
x=np.random.random((12, 12))           #heatmap of 12*12 matrix
plt.imshow(x)
plt.show()
```
